zip() function

```python
array_1 = [1, 2, 3, 4]
array_2 = [1, 2, 3, 4]

list(zip(array_1, array_2)) #returns a list of tuples with joined arrays
```

enumerate() function

```python
countries = ['Mexico', 'United States', 'Canada']

for index, country in enumerate(countries, start=1):
    print(f"Index: {index} Country: {country}")
```
