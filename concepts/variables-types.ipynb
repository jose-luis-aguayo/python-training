```python
x = 'str'
print(type(x)) #str

x = 1
print(type(x)) #int

x = 20.5
print(type(x)) #float

x = 1j
print(type(x)) #complex

x = ['apple', 'banana', 'cherry']
print(type(x)) #list

x = ('apple', 'banana', 'cherry')
print(type(x))

x = range(6)
print(type(x)) #range

x = {'name': 'jose', 'lastName': 'luis'}
print(type(x)) #dict

x = True
print(type(x)) #bool

x = None
print(type(x)) #NoneType
```
